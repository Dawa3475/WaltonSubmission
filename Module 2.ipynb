I pulled data of cereals and their nutritional values. I wanted to explore this data to find which cereals were the unhealthiest choices.  Seeing which cereals have the most sugars and calories and seeing sugary choice and contained the most calories.  

Comparing this data of the unhealthiest choices I wanted to see which of the unhealthy cereals were the most popular.  Doing this made me reflect on the cereal brands and what they portray to the public through ads and other platforms.  Are most of them being transparent about what their cereals contain?


```python
import pandas as pd
import altair as alt

data = pd.read_csv("cereal.csv")
data.head()
```

```python
alt.Chart(data).mark_circle().encode(
    x = "sugars",
    y = "calories",
    color=alt.Color('name', scale=alt.Scale(scheme='spectral')),
    size="rating",
    tooltip=["name", "calories"]
)
```

In order to see which cereals were the unhealthiest choice I wanted to see which ones contained the most sugars.  By creating an interactive scatter plot, I was able to see not only which cereals had the most sugars but also, which ones are the highest in calories and most popular cereals. After doing this I saw that Apply Jacks were one of the most popular sugary cereals, what surprised me was that Bran cereals like total raisin brand were among the most sugary and high caloric cereals.

```python
alt.Chart(data).mark_square().encode(
    x='sodium',
    y='calories',
    color=alt.Color('name', scale=alt.Scale(scheme='spectral')),
    size="rating",
    tooltip=["name", "calories"]
)
```

I then wanted to see which cereals were the highest in sodium as this is another important factor to the nutritional value of a cereal.  Rice Crispies and Corn Flakes were the cereals that have high levels of sodium.  However, it shocked me once again how high in sodium some of the bran cereals were.  These bran cereals are very popular as well.

```python
alt.Chart(data).mark_bar().encode(
    x="protein",
    y="name",
    color=alt.Color('name', scale=alt.Scale(scheme='spectral'))
)
```

Lastly, I wanted to see which cereals had the most protein as this is in contrast to the unhealthy theme I had been following.  By exploring which cereals are the highest in protein I can see which cereals are on the healthier, heartier side of the spectrum.  Cheerios and Special K were the highest in protein out of all the cereals.  The brans and oat cereals were relatively high in protein as well which made me think about cereals and what they contain as a whole.  It is important to note that the bran cereals were high in sugars and sodium, they are also on the higher end in proteins as well. Making Bran cereals the most controversial cereal.
